#Texto en Ruso a Voz

```python
!pip install gTTS    #Instala gTTS
from gtts import gTTS #Import Google Text to Speech
from IPython.display import Audio #Import Audio method from IPython's Display Class
```

```python
def escribir(texto,nombre):
    tts = gTTS(texto, lang='ru', slow=True) #Provide the string to convert to speech
    tts.save(nombre+'.mp3') #save the string converted to speech as a .wav file
    sound_file = nombre+'.mp3'
    return Audio(sound_file, autoplay=True)
print("Escribe el texto en Ruso")
texto = input()
print("Escribe el nombre del archivo")
nombre=input()
escribir(texto,nombre)
```
